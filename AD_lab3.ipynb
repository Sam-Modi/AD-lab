```python
#line plot
print("Sambhav Modi - 2205588")
import matplotlib.pyplot as plt

# Line plot
df['area'].plot(kind='line', title='Area Line Plot')
plt.xlabel('Index')
plt.ylabel('Area')
plt.grid(True)
plt.show()

# Histogram
print("Sambhav Modi - 2205588")
df['price'].plot(kind='hist', bins=10, title='Price Distribution')
plt.xlabel('Price')
plt.grid(True)
plt.show()
import seaborn as sns

# Scatter Plot (Matplotlib)
print("Sambhav Modi - 2205588")
plt.figure(figsize=(7, 3))
plt.scatter(df['area'], df['price'], color='blue', alpha=0.9)
plt.title('Scatter Plot: Area vs Price')
plt.xlabel('Area')
plt.ylabel('Price')
plt.grid(True)
plt.show()

# Histogram (Seaborn)
print("Sambhav Modi - 2205588")
plt.figure(figsize=(6, 4))
sns.histplot(df['price'], bins=5, kde=True, color='green')
plt.title('Price Distribution')
plt.xlabel('Price')
plt.show()

# Bar Chart (Seaborn)
print("Sambhav Modi - 2205588")
plt.figure(figsize=(6, 4))
sns.barplot(x='area', y='price', data=df, palette='viridis')
plt.title('Bar Chart: Price')
plt.xlabel('Index')
plt.ylabel('Price')
plt.show()

# Box Plot (Seaborn)
print("Sambhav Modi - 2205588")
plt.figure(figsize=(6, 4))
sns.boxplot(x='area', y='price', data=df, palette='coolwarm')
plt.title('Box Plot: Area vs Price')
plt.show()

# Heatmap (Seaborn)
print("Sambhav Modi - 2205588")
plt.figure(figsize=(6, 4))
correlation = df.corr()  # Compute the correlation matrix
sns.heatmap(correlation, annot=True, cmap='Blues')
plt.title('Correlation Heatmap')
plt.show()
```
